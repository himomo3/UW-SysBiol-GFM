```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
nsd = pd.read_csv("niche_summary_dist.csv")
nsd_gen = pd.read_csv("niche_summary_generation_dist.csv")
```

```python
nsd.head()
```

```python
nsd_gen.head()
```

```python
df = nsd_gen[nsd_gen['alpha0y'] != 0]

# Set up the figure with two subplots
fig, axes = plt.subplots(2, 1, figsize=(12, 10), sharex=True)

# Plot for amountniche1
sns.lineplot(ax=axes[0], data=df, x="generation_number", y="amountniche1", hue="alpha0y", palette="viridis")
axes[0].set_title("Individuals in Niche 1 over Generations")
axes[0].set_ylabel("Individuals in Niche 1")
axes[0].grid(True)
axes[0].legend(title="dist/2", bbox_to_anchor=(1.05, 1), loc='upper left')

# Plot for amountniche2
sns.lineplot(ax=axes[1], data=df, x="generation_number", y="amountniche2", hue="alpha0y", palette="viridis")
axes[1].set_title("Individuals in Niche 2 over Generations")
axes[1].set_xlabel("Generation Number")
axes[1].set_ylabel("Individuals in Niche 2")
axes[1].grid(True)
axes[1].legend(title="dist/2", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.savefig("fig1.png")
plt.show()
```

```python
nsms = pd.read_csv("niche_summary_mu_sigma.csv")
nsms_gen = pd.read_csv("niche_summary_generation_mu_sigma.csv")
```

```python
df1 = nsms.pivot(index="sigma", columns="mu", values="count_0")
df2 = nsms.pivot(index="sigma", columns="mu", values="count_1")

# Set up the figure with two subplots
fig, axes = plt.subplots(2, 1, figsize=(12, 10), sharex=True)
cmap = sns.light_palette("seagreen", as_cmap=True)
#cmap = "rocket"

# Plot for df1
sns.heatmap(ax=axes[0], data=df1, cmap=cmap)

# Plot for df2
sns.heatmap(ax=axes[1], data=df2, cmap=cmap)

plt.tight_layout()
plt.savefig("fig2.png")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Wczytanie danych z pliku CSV
df = pd.read_csv("niche_summary_alpha_xi.csv")  # Zmie na waciw ciek, jeli trzeba
df = df[df["alpha0y"]!=0]


#Tworzenie tabeli przestawnej dla count_0
pivot_0 = df.pivot_table(index='alpha0y', columns='xi', values='count_0')

#Tworzenie tabeli przestawnej dla count_1
pivot_1 = df.pivot_table(index='alpha0y', columns='xi', values='count_1')

#Rysowanie heatmapy dla count_0
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_0, cmap="YlGnBu", annot=True, fmt="g")
plt.title("Heatmapa count_0")
plt.xlabel("xi")
plt.ylabel("alpha0y")
plt.tight_layout()
plt.show()

#Rysowanie heatmapy dla count_1
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_1, cmap="YlOrRd", annot=True, fmt="g")
plt.title("Heatmapa count_1")
plt.xlabel("xi")
plt.ylabel("alpha0y")
plt.tight_layout()
plt.show()
```
